```python
# Load the pandas library
import pandas as pd

# Import into a DataFrame
df = pd.read_csv("Amazon_Consumer_Reviews_of_Amazon_Products.csv")

print(df.head(5))
```

```python
# 1st round of cleaning 
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda x: clean_text_round1(x)

df['reviews.text'] = df['reviews.text'].apply(round1)
df['reviews.text']
```

```python
df['reviews.title'] = df['reviews.title'].apply(round1)
print(df.head(5))
```

```python
#2nd round
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text

round2 = lambda x: clean_text_round2(x)
df['reviews.text'] = df['reviews.text'].apply(round2)
df['reviews.title'] = df['reviews.title'].apply(round2)
print(df.head(5))
```

```python
# Create a mapping based on the provided grouping
category_mapping = {
    'Animals & Pet Supplies': 'Animals & Pet Supplies',
    'Office Supplies': 'Office Supplies',
    'Office Supplies,Electronics': 'Electronics,Office Supplies',
    'Toys & Games,Electronics': 'Electronics,Toys & Games',
    'Health & Beauty': 'Electronics,Health & Beauty',
    'Electronics,Furniture': 'Electronics,Furniture',
    'Electronics': 'Electronics',
    'Electronics,Media': 'Electronics,Media',
    'Home & Garden': 'Home & Garden'
}

# Create the new 'Category' column based on the mapping
# If a category isn't in the mapping, use the original category
df['DCategory'] = df['primaryCategories'].map(category_mapping).fillna(df['primaryCategories'])
```

```python
from collections import Counter

# Preprocess the reviews
df['processed_reviews'] = df['reviews.text'].str.lower().str.translate(str.maketrans('', '', string.punctuation)).str.split()
df['processed_title'] = df['reviews.title'].str.lower().str.translate(str.maketrans('', '', string.punctuation)).str.split()

```

```python
# List of common stopwords to be excluded
stopwords = set(['the', 'i', 'to', 'and','my', 'very','after', 'but', 'these', 'around','could', 'so', 'a','over', 'never', 'amazon', 'in', 'for', 'is', 'on', 'with', 'as', 'it', 'at', 'by', 'an', 'be', 'this', 'that', 'of', 'was', 'they', 'are', 'use', 'all', 'its', 'if', 'not', 'she', 'just', 'though', 'then', 'no', 'their', 'were', 'should', 'will','am', 'too', 'had', 'while', 'however', 'does', 'about', 'your', 'would', 'really', 'her', 'we', 'not', 'have', 'do', 'also','from', 'has', 'them', 'me','one','or','which','because','there','when','only', 'some', 'out', 'did', 'you','can','got','they'])
```

```python
#Sentiment graph
# Filter out non-string reviews
#df = df[df['reviews.text'].apply(lambda x: isinstance(x, str))]
from textblob import TextBlob
import matplotlib.pyplot as plt
#You can see that the range of polarity varies between -1 and 1 where the positive end denotes positive tone, and negative end denotes negative tone in the sentence. 
#Hence, a polarity of 0 would be neutral. 
# Calculate sentiment using TextBlob's polarity
def textblob_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'
# Apply the TextBlob sentiment function to the filtered 'reviews.text' column
df['textblob_sentiment'] = df['reviews.text'].apply(textblob_sentiment)

def f(row):
    if row['reviews.rating'] >3:
        val = 'Positive'
    elif row['reviews.rating'] <3:
        val = 'Negative'
    else:
        val = 'Neutral'
    return val

df['rating'] = df.apply(f, axis=1)
# Get sentiment distribution for both methods
rating_sentiment_distribution = df['rating'].value_counts()
textblob_sentiment_distribution = df['textblob_sentiment'].value_counts()

# Plotting the sentiment distribution
fig, ax = plt.subplots(figsize=(12, 7))
width = 0.35
ind = range(len(rating_sentiment_distribution))

p1 = ax.bar(ind, rating_sentiment_distribution.values, width, label='Rating-based Sentiment')
p2 = ax.bar([i + width for i in ind], textblob_sentiment_distribution.values, width, label='TextBlob Sentiment')

ax.set_title('Sentiment Distribution Comparison')
ax.set_xticks([i + width / 2 for i in ind])
ax.set_xticklabels(rating_sentiment_distribution.index)
ax.legend()

plt.tight_layout()
plt.show()
from matplotlib import pyplot as plt

fig.savefig('polarity.png')
```

```python
#Word Cloud

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
# Using the STOPWORDS set from WordCloud library
stopwords_set = set(STOPWORDS)

# Concatenate all the reviews into one text after removing stopwords
#all_reviews = ' '.join([' '.join([word for word in review if word not in stopwords_set]) for review in df['reviews.text'].astype(str)])
my_reviews = df['processed_reviews'].apply(lambda x: [word for word in x if word not in stopwords])
#my_reviews = df['processed_reviews'].apply(lambda x: [word for word in x if word not in stopwords_set])
# Concatenate all the reviews into one text
all_reviews = ' '.join(my_reviews.astype(str))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, stopwords=stopwords, background_color='white').generate(all_reviews)

# Plot the word cloud
pcloud=plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Amazon Reviews')
plt.show()
# Save word cloud
from matplotlib import pyplot as plt

pcloud.savefig('wordcloud.png')
```

```python
# Bar graph
# Adjusting the code to rotate the primary categories' labels and remove grid lines
# Group the data by Categories
grouped_data = df.groupby('DCategory').agg(
    Count=('id', 'count'),
    AverageRating=('reviews.rating', 'mean')
).reset_index()

# Create the figure and axes objects
fig1, ax1 = plt.subplots(figsize=(12, 10))

# Plot the count of rows for each Categories
color = 'tab:blue'
ax1.set_xlabel('Categories')
ax1.set_ylabel('Count of Reviews', color=color)
bars = ax1.bar(grouped_data['DCategory'], grouped_data['Count'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(grouped_data['DCategory'], rotation=90)  # Rotate labels

# Adding the count of reviews on top of the bars
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', color=color)

# Create a twin Axes sharing the same x-axis
ax2 = ax1.twinx()  

# Plot the average reviews.rating for each Categories
color = 'tab:red'
ax2.set_ylabel('Average Rating', color=color)
ax2.plot(grouped_data['DCategory'], grouped_data['AverageRating'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# Title and layout
plt.title('Count and Average Rating of Reviews per Category')
fig1.tight_layout()

# Remove grid lines
ax1.grid(False)
ax2.grid(False)

# Show the plot
plt.show()
fig1.savefig('bar.png')
```
